```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_columns = None
```

```python
def extractCsv(sheet):
    dt = pd.read_csv(
        sheet, sep=',', header=None, 
        names=["price", "latitude", "longitude", "bedrooms", "bathrooms", "area", "pkspaces", "ensuites", "timestamp", "type", "operation", "url"])
    #print(dt.shape)
    dt = dt.dropna(axis='columns')
    #print(dt.head())
    return dt
```

```python
df_poa = extractCsv("POA_sample.csv")
df_poa

df_rj = extractCsv("RJ_sample.csv")
df_rj

df_sp = extractCsv("SP_sample.csv")
df_sp
```
